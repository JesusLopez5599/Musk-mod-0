```python
import requests

url_inicial = "https://jsonplaceholder.typicode.com/posts"

def obtener_posts():
    response = requests.get(url_inicial)
    if response.status_code == 200:
        print("Lista de posts obtenida con éxito:")
        print(response.json())  
    else:
        print(f"Error al obtener los posts. Código de estado: {response.status_code}")
```

```python
import requests

url_inicial = "https://jsonplaceholder.typicode.com/posts"

def crear_post():
    titulo = input("Ingresa el título del nuevo post: ")
    cuerpo = input("Ingresa el cuerpo del nuevo post: ")
    usuario_id = input("Ingresa el ID del usuario: ")

    data = {
        "title": titulo,
        "body": cuerpo,
        "userId": usuario_id
    }

    response = requests.post(url_inicial, json=data)
    if response.status_code == 201:
        print("Nuevo post creado con éxito:")
        print(response.json())
    else:
        print(f"Error al crear el post. Código de estado: {response.status_code}")
```

```python
import requests

url_inicial = "https://jsonplaceholder.typicode.com/posts"

def actualizar_post():
    titulo = input("Ingresa el título del nuevo post: ")
    cuerpo = input("Ingresa el cuerpo del nuevo post: ")
    usuario_id = input("Ingresa el ID del usuario: ")

    data = {
        "title": titulo,
        "body": cuerpo,
        "userId": usuario_id
    }
    
    response=requests.put(url_inicial, data)
    if response.status_code == 200:
        print("Post actualizado con éxito:")
        print(response.json())
    else:
        print(f"Error al crear el post. Código de estado: {response.status_code}")
```

```python
import requests

url_inicial = "https://jsonplaceholder.typicode.com/posts"

def patch_post(id, title):
    data= {"title": title}
    response =requests.patch(url_inicial/{id}, json=data)
    if response.status_code == 200:
        print("Título del post actualizado con éxito:")
        print(response.json())
    else:
        print(f"Error al crear el post. Código de estado: {response.status_code}")
```

```python
import requests

url_inicial = "https://jsonplaceholder.typicode.com/posts"

def delete_post(post_id):
    response = requests.delete(url_inicial/{post_id})
    if response.status_code == 200:
        print(f"Post con ID {post_id} eliminado exitosamente.")
    else:
        print(f"Error al eliminar el post: {response.status_code}")
```

Ejercicio 2.
Crear un programa en Python que interactúe con una API pública para obtener información
sobre el clima de una ciudad específica.
Utiliza la biblioteca requests de Python para enviar solicitudes HTTP a una API pública que
proporciona datos meteorológicos. La API que utilizaremos es OpenWeatherMap, que ofrece
datos meteorológicos actuales y pronósticos para ciudades de todo el mundo.

- Crea un script en Python que solicite al usuario el nombre de una ciudad.

- Utiliza la biblioteca requests para enviar una solicitud GET a la API del clima de
OpenWeatherMap, proporcionando el nombre de la ciudad y tu clave de API en la URL.

- Procesa la respuesta de la API y extrae la información relevante, como la temperatura actual
y la descripción del clima.

- Imprime la información del clima en la consola, incluyendo la descripción del clima y la
temperatura en grados Celsius.

```python
import requests

def obtener_clima(ciudad, api_key):
    url = f"https://api.openweathermap.org/data/2.5/weather?q={ciudad}&appid={api_key}&lang=es"
    try:
        # Realizar la solicitud GET
        respuesta = requests.get(url)
        respuesta.raise_for_status()  
        datos = respuesta.json()

        # Verificar si la ciudad es válida
        if datos["cod"] != 200:
            print(f"Error: {datos['message']}")
            return

        # Extraer información relevante
        descripcion = datos["weather"][0]["description"]
        temperatura_kelvin = datos["main"]["temp"]
        temperatura_celsius = temperatura_kelvin - 273.15

        # Mostrar resultados
        print(f"El clima en {ciudad} es: {descripcion}")
        print(f"Temperatura: {temperatura_celsius:.2f} grados Celsius")

    except requests.exceptions.RequestException as e:
        print(f"Error al realizar la solicitud: {e}")
    except KeyError:
        print("Error al procesar los datos recibidos de la API.")

if __name__ == "__main__":
    ciudad = input("Introduce el nombre de la ciudad: ")
    api_key= input("Introduce tu codigo personal:")
    obtener_clima(ciudad, api_key)
```
