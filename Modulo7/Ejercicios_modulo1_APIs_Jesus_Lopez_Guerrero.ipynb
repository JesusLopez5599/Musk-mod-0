Recibe un límite inferior y superior para el precio y muestra todos los libros cuyo precio esta
en ese intervalo.

```python
import json

def filtrar_libros(file_path, lim_inf, lim_sup):
    with open(file_path, 'r') as file:
      data = json.load(file)
    
    libros = data['bookstore']['book']
    libros_en_rango=[
        libro for libro in libros
        if lim_inf <= float(libro['price']) <= lim_sup
                     ]  
    return libros_en_rango

lim_inferior=float(input("Introduce un límite inferior:"))
lim_superior=float(input("Introduce un límite superior:"))

libros_filtrados = filtrar_libros('ej1.json', limite_inferior, limite_superior)

print("Libros en el rango de precio especificado:")
for libro in libros_filtrados:
  print(f" Título:{libro['title']['__text']}, Precio: {libro['price']}")
```

Actualizar el precio de un libro dado su título. Implementa una función que reciba el título de
un libro y un nuevo precio, y actualice el precio del libro en el archivo JSON.

```python
import json

def actualizar_precio(file_path, titulo, precio):
    with open(file_path,'r') as file:
        data = json.load(file)
    libros= data ['bookstore']['book']
    libro_encontrado=False
    for libro in libros:
        if libro['title']['__text'].lower()==titulo.lower():
            libro['price']=str(precio)
            libro_encontrado=True
            break
    if libro_encontrado:
        with open(file_path,'w') as file:
            json.dump(data,file, indent=4)
            print(f"El precio del libro '{titulo} se ha actualizado a '{precio}'.")
    else:
        print(f"No se encontró un libro con el título '{titulo}'.")
titulo_libro= input("introduce el título del libro:")
nuevo_precio= float(input("introduce el nuevo precio:"))
    
actualizar_precio('ej1.json', titulo_libro, nuevo_precio)    
```

¿Qué libros tienen un precio por debajo del promedio de todos los precios?

```python
def libros_por_debajo_del_promedio(file_path):
    with open(file_path, 'r') as file:
        data = json.load(file)
    libros = data['bookstore']['book']
    
    precios = [float(libro['price']) for libro in libros]
    precio_promedio = sum(precios) / len(precios)
    
    libros_bajo_promedio = [
        {"título": libro['title']['__text'], "precio": float(libro['price'])}
        for libro in libros
        if float(libro['price']) < precio_promedio
    ]
    
    return libros_bajo_promedio, precio_promedio

resultado, promedio = libros_por_debajo_del_promedio('ej1.json')

print(f"El precio promedio es {promedio:.2f}.")
if resultado:
    print("Libros con precio por debajo del promedio:")
    for libro in resultado:
        print(f"- Título: {libro['título']}, Precio: {libro['precio']:.2f}")
else:
    print("No hay libros con precio por debajo del promedio.")
```

Recibe una fecha y devuelve las pruebas que comienzan o terminan en esa fecha.
Implementa una función que reciba una fecha y devuelva una lista de pruebas que
empiezan o terminan en esa fecha.

```python
import json
from datetime import datetime

with open('ej2.json','r',encoding='utf-8') as file:
    data =json.load(file)

def buscar_prueba_por_fecha(data, fecha_busqueda):
    fecha_busqueda=datetime.strptime(fecha_busqueda, "%Y-%m-%d").date()

    pruebas_encontradas = [
        prueba['Titulo']
        for prueba in data
        if datetime.strptime(prueba['InicioImparticion'], "%Y-%m-%dT%H:%M:%S").date() == fecha_busqueda or
           datetime.strptime(prueba['FinImparticion'], "%Y-%m-%dT%H:%M:%S").date() == fecha_busqueda
    ]
    return pruebas_encontradas

fecha_input = input("Introduce la fecha:")
pruebas_resultado = buscar_pruebas_por_fecha(data, fecha_input)

if pruebas_resultado:
    print(f"Pruebas que comienzan o terminan el {fecha_input}:")
    for prueba in pruebas_resultado:
        print(f"- {prueba}")
else:
    print(f"No se encontraron pruebas que comiencen o terminen el {fecha_input}.")
```

Cuántas provincias y municipios hay en total? Implementa una función que devuelva el
número total de provincias y el número total de municipios.

```python
import json

with open('ej3.json', 'r', encoding='utf-8') as file:
    data = json.load(file)

def contar_provincias_y_municipios(data):
    total_provincias = len(data['lista']['provincia'])
    total_municipios = sum(len(provincia['localidades']['localidad']) for provincia in data['lista']['provincia'])
    return total_provincias, total_municipios

total_provincias, total_municipios = contar_provincias_y_municipios(data)

print(f"Total de provincias: {total_provincias}")
print(f"Total de municipios: {total_municipios}")
```
