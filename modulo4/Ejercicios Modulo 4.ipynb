```python
from abc import ABC, abstractmethod

class Model(ABC):
    @abstractmethod
    def preprocess_data(self, data=None, y=None):
        pass

    @abstractmethod
    def fit(self):
        pass

    @abstractmethod
    def predict(self):
        pass

class SVM(Model):
    def preprocess_data(self, data=None, y=None):
        print("Preprocessing data at SVM")

    def fit(self):
        print("Training at SVM")

    def predict(self):
        print("Evaluating at SVM")

class DecisionTree(Model):
    def preprocess_data(self, data=None, y=None):
        print("Preprocessing data at DecisionTree")

    def fit(self):
        print("Training at DecisionTree")

    def predict(self):
        print("Evaluating at DecisionTree")

#Ejemplo
svm=SVM()
svm. preprocess_data(data=None, y=None)
svm.fit()
svm.predict()

dt=DecisionTree()
dt. preprocess_data(data=None, y=None)
dt.fit()
dt.predict()
```

Ejercicio 14. Escribe un programa en Python para seleccionar todos los domingos de un año determinado

```python
from datetime import datetime, timedelta

def Obtener_Domingos(year):
    #iniciar con el 1 de enero
    fecha_inicial=datetime(year,1,1)

    #lista vacía que acumula los domingos
    Domingos= []

    #construimos el bucle que iterará cada día del año
    while fecha_inicial.year ==year:
        if fecha_inicial.weekday()==6:
            Domingos.append(fecha_inicial.strftime('%Y-%m-%d'))
        fecha_inicial += timedelta(days=1)
    
    return Domingos

#Ejemplo 
year =2024
domingos=Obtener_Domingos(year)

print(f"Todos los domingos de {year}:")
for domingo in domingos:
    print(domingo)
```

Ejercicio 15. Escribe un programa en Python para contar el número de lunes del primer día del mes desde 2015 hasta 2016.

```python
from datetime import datetime

def Obtener_Lunes_1(año_inicial, año_final):
    numero_Lunes=0

    for year in range(año_inicial, año_final +1):
        for month in range(1,13):
            fecha = datetime(year,month,1)
            if fecha.weekday() ==0:
                numero_Lunes +=1
    
    return numero_Lunes

#ejemplo
año_inicial=2015
año_final=2016
num_Lunes= Obtener_Lunes_1(año_inicial,año_final)

print (f"El número de lunes que cayeron día 1 desde el año {año_inicial} hasta {año_final} fueron de {num_Lunes}")
```

Ejercicio 16. Escribe un programa en Python para crear 12 fechas fijas a partir de una fecha especificada en un periodo determinado. La diferencia entre dos fechas será de 20

```python
from datetime import datetime,timedelta

def Crea_fechas_fijas(fecha_inicial, numero, Intervalo):
    fechas=[]
    fecha= datetime.strptime(fecha_inicial, "%Y-%m-%d")

    for _ in range(numero):
        fechas.append(fecha.strftime("%Y-%m-%d"))
        fecha +=timedelta(days=Intervalo)

    return fechas

#Fecha inicial especificada
fecha_inicial = "2021-05-05"
# Número de fechas a generar
numero_fechas = 12
# Intervalo de días entre cada fecha
intervalo_dias = 20

# Crear las fechas fijas
fechas = Crea_fechas_fijas(fecha_inicial, numero_fechas, intervalo_dias)

# Mostrar las fechas generadas
print("Fechas generadas:")
for fecha in fechas:
    print(fecha)
```

Ejercicio 17. Implementa una función generadora que dadas dos listas del mismo tamaño, devuelva la multiplicación entre los elementos de cada una. Añadiendo el bloque except capturamos la excepción de Stop Iteration que se produce al acabar de leer todos los elementos de un generador.

```python
def prod(l1,l2):
    # verificamos que tengan el mismo tamaño
    if len(l1)!=len(l2):
        raise ValueError("Las listas deben tener el mismo tamaño")

    for a, b in zip(l1,l2):
        yield a * b
# Listas de ejemplo
lista1 = [1, 2, 3, 4, 5]
lista2 = [6, 7, 8, 9, 10]

generador= prod(lista1,lista2)
while True:
    try:
        resultado = next(generador)
        print(f"Resultado de la multiplicación: {resultado}")
    except StopIteration:
        # Capturar la excepción StopIteration cuando el generador se ha agotado
        print("Todos los elementos del generador han sido procesados.")
        break
```

Ejercicio 18. Implementa un generador, que dado un entero n, genere n número aleatorios. Puedes usar el método random de la librería random para generar números aleatorios.

```python
import random
def generador_random(n):
    for _ in range(n):
        yield random.random()

n=12

generador= generador_random(n)
print("Números aleatorios generados:")
for numero in generador:
    print(numero)
```

Ejercicio 19. Implementa uun generador de Fibonacci que genere n números de la secuencia de Fibonacci, que tiene la forma: 0,1,1,2,3,5,...
Cuyos dos primeros valores son 0 y 1 por definición y el resto se calcular sumando los dos últimos valores de la sucesión

```python
def sucesión_F (n):
    a,b =0, 1
    for _ in range(n):
        yield a
        a,b = b, a+b

n=20

generador=sucesión_F(n)
print("Números de la secuencia de Fibonacci generados:")
for numero in generador:
    print(numero)
```

Ejercicio 23. Crea una función que compute la diferencia entre dos enteros. En caso de que la diferencia sea negativa genera una excepción inventada por ti que informe sobre ello. Por ejemplo, la excepción podría llamarse NegativeDifferenceException

```python
# Definición de la excepción personalizada
class NegativeDifferenceException(Exception):
    def __init__(self, difference, message="La diferencia es negativa"):
        self.difference = difference
        self.message = message
        super().__init__(self.message)

    def __str__(self):
        return f"{self.message}: {self.difference}"

# Función para calcular la diferencia entre dos enteros
def calcular_diferencia(a, b):
    diferencia = a - b
    if diferencia < 0:
        raise NegativeDifferenceException(difference=diferencia)
    return diferencia

# Prueba de la función con manejo de excepciones
try:
    resultado = calcular_diferencia(5, 10)
    print(f"La diferencia es: {resultado}")
except NegativeDifferenceException as e:
    print(f"Se ha producido una excepción: {e}")

# Otro ejemplo sin excepción
try:
    resultado = calcular_diferencia(10, 5)
    print(f"La diferencia es: {resultado}")
except NegativeDifferenceException as e:
    print(f"Se ha producido una excepción: {e}")
```

Ejercicio 24. Crea una función que calcule la división entre dos números. Debe imprimir el mensaje 'Los parámetros deben ser número enteros' cuando se genera una excepción de tipo y ' El divisor no puede ser 0' cuando se genera un zerodivisionerror.

```python
def dividir_numeros(a,b):
    try:
        c=a/b
        return c
    except TypeError:
        print("Los parámetros deben ser números enteros")
    except ZeroDivisionError:
        print("El divisor no puede ser 0")

print(dividir_numeros(10, "2"))
```
