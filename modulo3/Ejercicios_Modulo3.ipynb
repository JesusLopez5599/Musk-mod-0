7. Añade un método público en la clase Estudiante que calcule la media de una lista de notas y actualice el valor del atributo grade. A continuación llama a la función en tu programa principal e imprime el valor de tu grade

```python
class Estudiante:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad
        self.grade = None

    def calcular_media(self, notas):
        if notas:
            self.grade = sum(notas) / len(notas)
        else:
            self.grade = 0
    
    def informacion (self):
        return "Nombre: {self.nombre}, Edad: {self.edad}, Grado: {self.grade}"

# Crear un objeto de la clase Estudiante
estudiante1 = Estudiante("Juan Pérez", 20)

# Lista de notas
notas = [85, 90, 78, 92, 88]

# Calcular la media y actualizar el atributo grade
estudiante1.calcular_media(notas)

# Mostrar la información del estudiante
print(estudiante1.informacion())
```

10. Añade un método privado de la clase anterior, que dado un diccionario mes-número de asistencias, devuelva 1 si un mes tiene menos de 4 asistencias, devuelva 2 si algún mes tiene entre [4,8) asistencias o 3 en caso contrario. Para probar el resultado, encapsúlalo con una función pública que devuelva su resultado.

```python
class Estudiante:

    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

    
    @classmethod
    def actualizar_escuela(cls, nueva_escuela):
        cls.escuela = nueva_escuela

    def __evaluar_asistencias(self, asistencias):
        for mes, numero in asistencias.items():
            if numero < 4:
                return 1
            elif 4 <= numero < 8:
                return 2
        return 3

    def evaluar_asistencias_publica(self, asistencias):
        return self.__evaluar_asistencias(asistencias)


# Crear instancias de Estudiante
estudiante1 = Estudiante("Juan Perez", 20)

# Diccionario de asistencias 
asistencias = {
    "Septiembre": 6,
    "Octubre":2,
    "Noviembre":10,
    "Diciembre":6,
    "Enero": 3,
    "Febrero": 5,
    "Marzo": 8,
    "Abril": 4,
    "Mayo": 2
}


resultado_asistencias = estudiante1.evaluar_asistencias_publica(asistencias)
print("Resultado de la evaluación de asistencias:", resultado_asistencias)
    
```
