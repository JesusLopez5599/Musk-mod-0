Ejercicio 1. Escribe una función en python para leer el contenido de un archivo de texto "poema.txt" línea por línea y moostrar el mismo en pantalla.

```python
def leer_poema ():
    try:
        with open("poema.txt", "r", encoding="utf-8") as archivo:
            lista_lineas=archivo.readlines()
            for linea in lista_lineas:
                print(linea.readline())
    except FileNotFoundError:
        print("El archivo 'poema.txt' no se encontró.")
        
```

Ejercicio 2: Escribe una función para contar el número de líneas de un archivo de texto "historia.txt".

```python
def con_lineas():
    try:
        with open("historia.txt","r", encoding="utf-8") as archivo:
            lista_lineas=archivo.readlines()
            n= len(lista_lineas)
            print (f"El archivo 'historia.txt' tiene {n} líneas.")
            return n
    except FileNotFoundError:
        print("El archivo 'historia.txt' no se encontró.")
        return 0
```

Ejercicio 4: Escriba una función en Python para leer líneas de un archivo de "notas.txt". Su función debe encontrar y mostrar la aparición de la palabra "el".

```python
def encontrar_palabra_el():
    try:
        with open("notas.txt", "r", encoding="utf-8") as archivo:
            lineas =archivo.readlines()
            for numero_linea, linea in enumerate(lineas, start=1):
                if "el" in linea.split():
                    print(f"La palabra 'el' se encontró en la línea {numero_linea}: {linea.strip()}")
    except FileNotFoundError:
        print("El archivo 'notas.txt' no se encontró.")
            
```

Ejercicio 5. Escriba una función display_words() en python para leer las lineas de un archivo de texto "story.txt", y mostrar aquellas palabras que tengan menos de 4 caracteres.

```python
def display_words():
    try:
        with open("story.ttxt", "r", encoding="utf-8") as archivo:
            lineas=archivo.readlines()
            for linea in lineas:
                palabras=linea.split()
                for palabra in palabras:
                    if len(palabra)<4:
                        print(palabra)
    except FileNotFoundError:
        print("El archivo 'story.txt' no se encuentra.")
```

Ejercicio 9. Escribe un programa en python para calcular la frecuencia de todas las palabras de un archivo txt.

```python
from collections import Counter
import re

def contar_palabras(archivo):
    try:
        with open(archivo, 'r', encoding='utf-8') as file:
            texto = file.read()
            texto1=texto.lower
            palabras=re.findall(r'\b\w+\b', texto1)
#La linea anterior quita signos de puntuación
            frecuencia_palabras= Counter(palabras)
            return frecuencia_palabras
    except FileNotFoundError:
        print("El archivo 'story.txt' no se encuentra.")
```

Ejercicio 39. Lee todos los datos de las ventas de productos y muéstrelos mediante el digrama de pila.

```python
import matplotlib.pyplot as plt
import pandas as pd

file_path='Modulo5_company_sales_data-221102-123259.csv'
data=pd.read_csv(file_path)

months = data['month_number']
facecream_sales = data['facecream']
facewash_sales = data['facewash']
toothpaste_sales = data['toothpaste']
bathingsoap_sales = data['bathingsoap']
shampoo_sales = data['shampoo']
moisturizer_sales = data['moisturizer']

plt.figure(figsize=(10, 6))

plt.bar(months, facecream_sales, label='Crema facial')
plt.bar(months, facewash_sales, bottom=facecream_sales, label='Lavado de cara')
plt.bar(months, toothpaste_sales, bottom=facecream_sales+facewash_sales, label='Pasta de dientes')
plt.bar(months, bathingsoap_sales, bottom=facecream_sales+facewash_sales+toothpaste_sales, label='Jabón de baño')
plt.bar(months, shampoo_sales, bottom=facecream_sales+facewash_sales+toothpaste_sales+bathingsoap_sales, label='Champú')
plt.bar(months, moisturizer_sales, bottom=facecream_sales+facewash_sales+toothpaste_sales+bathingsoap_sales+shampoo_sales, label='Hidratante')

plt.xlabel('Número de mes')
plt.ylabel('Número de unidades vendidas')
plt.title('Ventas mensuales de productos en un stack plot')

plt.legend(loc='upper left')

plt.show()
```
